```python
import os
import pathlib
import tempfile

import numpy as np
import pandas as pd
import pytest

from spyglass.utils.statescript import (
    StateScriptLogProcessor,
    _parse_int,
    parse_statescript_line,
    parse_ts_int_int,
    parse_ts_str,
    parse_ts_str_equals_int,
    parse_ts_str_int,
)


@pytest.fixture(scope="module")
def sample_log_content():
    """Provides sample log content."""
    return """# Test log
76504 0 0
76566 center_poke
76566 65536 0
100078 counter_handlePoke = 1
100078 4 0
100559 0 0
Executing this line
115030 center_poke
115030 65536 0
"""


@pytest.fixture
def processor(sample_log_content):
    """Provides a processor instance initialized with sample content."""
    return StateScriptLogProcessor(sample_log_content)


@pytest.fixture(scope="module")
def external_times():
    """Provides sample external times."""
    return np.array([1678886476.530, 1678886500.100, 1678886515.050])


@pytest.fixture
def temp_log_file(sample_log_content):
    """Creates a temporary log file and yields its path."""
    with tempfile.NamedTemporaryFile(
        mode="w", delete=False, suffix=".stateScriptLog"
    ) as tmp_file:
        tmp_file.write(sample_log_content)
        tmp_file_path = tmp_file.name
    yield tmp_file_path
    os.remove(tmp_file_path)


# --- Tests for Level 1 Parsers ---


def test_parse_int():
    """Test the _parse_int helper function."""
    assert _parse_int("123") == 123
    assert _parse_int("-45") == -45
    assert _parse_int("0") == 0
    assert _parse_int("abc") is None
    assert _parse_int("12.3") is None
    assert _parse_int("") is None
    assert (
        _parse_int(" 123 ") == 123
    )  # Should handle surrounding whitespace if not stripped before


def test_parse_ts_int_int_direct():
    """Test parse_ts_int_int directly."""
    parts = ["8386500", "0", "0"]
    expected = {
        "type": "ts_int_int",
        "timestamp": 8386500,
        "value1": 0,
        "value2": 0,
    }
    assert parse_ts_int_int(parts) == expected

    parts_wrong_len = ["123", "0"]
    assert parse_ts_int_int(parts_wrong_len) is None

    parts_not_int = ["123", "abc", "0"]
    assert parse_ts_int_int(parts_not_int) is None

    parts_float = ["123", "4.5", "0"]
    assert parse_ts_int_int(parts_float) is None


def test_parse_ts_str_int_direct():
    """Test parse_ts_str_int directly."""
    parts = ["8386500", "DOWN", "3"]
    expected = {
        "type": "ts_str_int",
        "timestamp": 8386500,
        "text": "DOWN",
        "value": 3,
    }
    assert parse_ts_str_int(parts) == expected

    parts_wrong_len = ["123", "UP"]
    assert parse_ts_str_int(parts_wrong_len) is None

    parts_str_is_int = ["123", "456", "789"]
    assert (
        parse_ts_str_int(parts_str_is_int) is None
    )  # Should be handled by ts_int_int

    parts_val_not_int = ["123", "UP", "abc"]
    assert parse_ts_str_int(parts_val_not_int) is None


def test_parse_ts_str_equals_int_direct():
    """Test parse_ts_str_equals_int directly."""
    parts = ["100078", "counter_handlePoke", "=", "1"]
    expected = {
        "type": "ts_str_equals_int",
        "timestamp": 100078,
        "text": "counter_handlePoke",
        "value": 1,
    }
    assert parse_ts_str_equals_int(parts) == expected

    parts_multi_word = ["3610855", "total", "rewards", "=", "70"]
    expected_multi = {
        "type": "ts_str_equals_int",
        "timestamp": 3610855,
        "text": "total rewards",
        "value": 70,
    }
    assert parse_ts_str_equals_int(parts_multi_word) == expected_multi

    parts_wrong_len = ["123", "=", "1"]
    assert parse_ts_str_equals_int(parts_wrong_len) is None

    parts_no_equals = ["123", "text", "1"]
    assert parse_ts_str_equals_int(parts_no_equals) is None

    parts_val_not_int = ["123", "text", "=", "abc"]
    assert parse_ts_str_equals_int(parts_val_not_int) is None


def test_parse_ts_str_direct():
    """Test parse_ts_str directly."""
    parts = ["76566", "center_poke"]
    expected = {"type": "ts_str", "timestamp": 76566, "text": "center_poke"}
    assert parse_ts_str(parts) == expected

    parts_multi_word = [
        "1271815",
        "lastPort",
        "=",
        "-1",
        "to",
        "currPort",
        "=",
        "2",
    ]
    expected_multi = {
        "type": "ts_str",
        "timestamp": 1271815,
        "text": "lastPort = -1 to currPort = 2",
    }
    assert parse_ts_str(parts_multi_word) == expected_multi

    parts_wrong_len = ["123"]
    assert parse_ts_str(parts_wrong_len) is None

    parts_second_is_int = [
        "123",
        "456",
    ]  # Second part is int, should fail this parser
    assert parse_ts_str(parts_second_is_int) is None


# --- Tests for parse_statescript_line (Covers integration and dispatching) ---


def test_parse_statescript_line_ts_int_int():
    """Test parse_statescript_line dispatching for ts_int_int."""
    line = "8386500 0 0"
    parsed = parse_statescript_line(line)
    assert parsed["type"] == "ts_int_int"
    assert parsed["timestamp"] == 8386500
    assert parsed["value1"] == 0
    assert parsed["value2"] == 0
    assert parsed["raw_line"] == line


def test_parse_statescript_line_ts_str_int():
    """Test parse_statescript_line dispatching for ts_str_int."""
    line = "8386500 DOWN 3"
    parsed = parse_statescript_line(line)
    assert parsed["type"] == "ts_str_int"
    assert parsed["timestamp"] == 8386500
    assert parsed["text"] == "DOWN"
    assert parsed["value"] == 3
    assert parsed["raw_line"] == line


def test_parse_statescript_line_ts_str_equals_int():
    """Test parse_statescript_line dispatching for ts_str_equals_int."""
    line = "100078 counter_handlePoke = 1"
    parsed = parse_statescript_line(line)
    assert parsed["type"] == "ts_str_equals_int"
    assert parsed["timestamp"] == 100078
    assert parsed["text"] == "counter_handlePoke"
    assert parsed["value"] == 1
    assert parsed["raw_line"] == line


def test_parse_statescript_line_ts_str():
    """Test parse_statescript_line dispatching for ts_str."""
    line = "76566 center_poke"
    parsed = parse_statescript_line(line)
    assert parsed["type"] == "ts_str"
    assert parsed["timestamp"] == 76566
    assert parsed["text"] == "center_poke"
    assert parsed["raw_line"] == line


def test_parse_statescript_line_unknown():
    """Test parse_statescript_line dispatching for unknown lines."""
    line = "Executing trigger function 22"  # No timestamp
    parsed = parse_statescript_line(line)
    assert parsed["type"] == "unknown"
    assert "timestamp" not in parsed
    assert parsed["raw_line"] == line


def test_parse_statescript_line_comment_empty():
    """Test parse_statescript_line dispatching for comments/empty."""
    line_c = "# comment"
    line_e = ""
    line_s = "   "
    assert parse_statescript_line(line_c)["type"] == "comment_or_empty"
    assert parse_statescript_line(line_c)["raw_line"] == line_c
    assert parse_statescript_line(line_e)["type"] == "comment_or_empty"
    assert parse_statescript_line(line_e)["raw_line"] == line_e
    assert parse_statescript_line(line_s)["type"] == "comment_or_empty"
    assert parse_statescript_line(line_s)["raw_line"] == ""


# --- Tests for StateScriptLogProcessor ---


def test_init_from_string(processor, sample_log_content):
    assert processor.log_content == sample_log_content
    assert processor.source_description == "from string"
    assert processor.raw_events == []
    assert processor.time_offset is None
    assert processor.processed_events_df is None


def test_init_from_file(temp_log_file, sample_log_content):
    processor_file = StateScriptLogProcessor.from_file(temp_log_file)
    assert processor_file.log_content == sample_log_content
    assert processor_file.source_description.startswith("from file:")
    assert pathlib.Path(temp_log_file).name in processor_file.source_description


def test_init_from_file_not_found():
    with pytest.raises(FileNotFoundError):
        StateScriptLogProcessor.from_file("non_existent_file.log")


def test_parse_raw_events(processor):
    events = processor.parse_raw_events()
    assert isinstance(events, list)
    assert len(events) == 10
    assert events[0]["type"] == "comment_or_empty"
    assert events[1]["type"] == "ts_int_int"
    assert events[7]["type"] == "unknown"
    assert events[1]["raw_line"] == "76504 0 0"
    assert events[7]["raw_line"] == "Executing this line"


def test_find_reference_events(processor):
    ref_df = processor._find_reference_events(
        event_type="ts_str", conditions={"text": "center_poke"}
    )
    assert isinstance(ref_df, pd.DataFrame)
    assert len(ref_df) == 2
    pd.testing.assert_series_equal(
        ref_df["timestamp"],
        pd.Series([76566, 115030], name="timestamp"),
        check_dtype=False,
    )
    assert "log_timestamp_sec" in ref_df.columns

    ref_df_num = processor._find_reference_events(
        event_type="ts_int_int", conditions={"value1": 4, "value2": 0}
    )
    assert len(ref_df_num) == 1
    assert ref_df_num["timestamp"].iloc[0] == 100078

    ref_df_none = processor._find_reference_events(
        event_type="ts_str_int", conditions={"text": "nonexistent"}
    )
    assert ref_df_none.empty


def test_calculate_time_offset_success(processor):
    ext_times = np.array([1678880076.566, 1678880115.030])
    offset = processor.calculate_time_offset(
        external_reference_times=ext_times,
        log_event_type="ts_int_int",
        log_event_conditions={"value1": 65536, "value2": 0},
        check_n_events=2,
    )
    assert offset is not None
    assert offset == pytest.approx(1678880000.0)


def test_calculate_time_offset_fail_not_enough_log(processor, external_times):
    offset = processor.calculate_time_offset(
        external_reference_times=external_times,
        log_event_type="ts_str_equals_int",
        log_event_conditions={"text": "counter_handlePoke"},
        check_n_events=2,
    )
    assert offset is None
    assert processor.time_offset is None


def test_calculate_time_offset_fail_not_enough_external(processor):
    offset = processor.calculate_time_offset(
        external_reference_times=np.array([1678880076.566]),
        log_event_type="ts_int_int",
        log_event_conditions={"value1": 65536, "value2": 0},
        check_n_events=2,
    )
    assert offset is None
    assert processor.time_offset is None


def test_get_events_dataframe_defaults(processor):
    """Test default behavior: exclude comments/unknown, no offset applied yet."""
    df = processor.get_events_dataframe(
        apply_offset=False
    )  # Default exclude=True
    assert isinstance(df, pd.DataFrame)
    assert len(df) == 8  # Excludes comment and unknown line
    assert "raw_line" in df.columns
    assert "timestamp" in df.columns
    assert "log_timestamp_sec" in df.columns
    # Check column order: time first, raw_line last
    expected_cols = [
        "timestamp",
        "log_timestamp_sec",
        "timestamp_sync",
        "text",
        "value",
        "value1",
        "value2",
        "raw_line",
        "type",
    ]
    actual_expected_cols = [col for col in expected_cols if col in df.columns]
    assert list(df.columns) == actual_expected_cols
    # Check content
    assert df["raw_line"].iloc[0] == "76504 0 0"
    assert pd.isna(df["text"].iloc[0])  # Should be NA where not applicable
    assert df["value1"].iloc[0] == 0


def test_get_events_dataframe_include_all(processor):
    """Test including comments and unknown lines."""
    df = processor.get_events_dataframe(
        apply_offset=False, exclude_comments_unknown=False
    )
    assert isinstance(df, pd.DataFrame)
    assert len(df) == 10  # Includes comment and unknown line
    assert "raw_line" in df.columns
    # Check raw_line for the unknown line
    assert (
        df["raw_line"].iloc[7] == "Executing this line"
    )  # Index adjusted for comment
    # Check that timestamp is NA or 0 for lines without one
    assert (
        pd.isna(df["timestamp"].iloc[0]) or df["timestamp"].iloc[0] == 0
    )  # Comment line
    assert (
        pd.isna(df["timestamp"].iloc[7]) or df["timestamp"].iloc[7] == 0
    )  # Unknown line
    # Check column order
    expected_cols = [
        "timestamp",
        "log_timestamp_sec",
        "timestamp_sync",
        "text",
        "value",
        "value1",
        "value2",
        "raw_line",
        "type",
    ]
    actual_expected_cols = [col for col in expected_cols if col in df.columns]
    assert list(df.columns) == actual_expected_cols


def test_get_events_dataframe_with_offset(processor):
    """Test applying offset and column order."""
    processor.time_offset = 1678880000.0
    df = processor.get_events_dataframe(
        apply_offset=True
    )  # Default exclude=True
    assert isinstance(df, pd.DataFrame)
    assert len(df) == 8
    # Check calculation
    expected_sync_time = (76504 / 1000.0) + 1678880000.0
    assert df["timestamp_sync"].iloc[0] == pytest.approx(expected_sync_time)
    # Check NA value handling
    assert pd.isna(df["text"].iloc[0])


test_parse_int()
test_parse_ts_int_int_direct()
test_parse_ts_str_int_direct()
test_parse_ts_str_equals_int_direct()
test_parse_ts_str_direct()
test_parse_statescript_line_ts_int_int()
test_parse_statescript_line_ts_str_int()
test_parse_statescript_line_ts_str_equals_int()
test_parse_statescript_line_ts_str()
test_parse_statescript_line_unknown()
test_parse_statescript_line_comment_empty()
test_parse_statescript_line_comment_empty()

sample_log_content = """# Test log
76504 0 0
76566 center_poke
76566 65536 0
100078 counter_handlePoke = 1
100078 4 0
100559 0 0
Executing this line
115030 center_poke
115030 65536 0
"""
processor = StateScriptLogProcessor(sample_log_content)
with tempfile.NamedTemporaryFile(
    mode="w", delete=False, suffix=".stateScriptLog"
) as temp_log_file:
    temp_log_file.write(sample_log_content)
    temp_log_file_path = temp_log_file.name

external_times = np.array([1678886476.530, 1678886500.100, 1678886515.050])

test_init_from_string(processor, sample_log_content)
test_init_from_file(temp_log_file_path, sample_log_content)
test_init_from_file_not_found()
test_parse_raw_events(processor)
test_find_reference_events(processor)
test_calculate_time_offset_success(processor)
test_calculate_time_offset_fail_not_enough_log(processor, external_times)
test_calculate_time_offset_fail_not_enough_external(processor)
test_get_events_dataframe_defaults(processor)
test_get_events_dataframe_include_all(processor)
test_get_events_dataframe_with_offset(processor)
```

```python
from typing import List, Optional
import numpy as np
import pandas as pd

def _interpret_port_mask(
    port_state_value: Optional[int], max_ports: int = 32
) -> List[int]:
    """
    Interprets an integer value as a bitmask representing active ports using NumPy.

    Assumes a 1-based port numbering system (e.g., bit 0 corresponds to port 1).

    Parameters
    ----------
    port_state_value : Optional[int]
        The integer value representing the combined state of multiple ports.
        Handles None or pandas NA values.
    max_ports : int, optional
        The maximum port number to check (bits 0 to max_ports-1), by default 32.

    Returns
    -------
    List[int]
        A sorted list of 1-based port numbers that are active (bit is set).
        Returns an empty list if the value is 0, None, or NA.

    Example
    -------
    >>> interpret_port_mask(9) # 1001 in binary -> Ports 1 and 4
    [1, 4]
    >>> interpret_port_mask(65536) # 2^16 -> Port 17
    [17]
    """
    # Return empty list for 0, None, or pandas NA
    if pd.isna(port_state_value) or port_state_value == 0:
        return []

    # Ensure value is treated as an integer after NA check
    try:
        port_state_int = int(port_state_value)
    except (ValueError, TypeError):
        # Should not happen if input is from Int64Dtype column after NA check,
        # but included for robustness if called directly with invalid input.
        return []

    # Create bit masks for positions 0 to max_ports-1
    # E.g., [1, 2, 4, 8, ...]
    bit_masks = np.left_shift(1, np.arange(max_ports))

    # Check which bits are set in the input value using bitwise AND
    active_bits_mask = np.bitwise_and(port_state_int, bit_masks) > 0

    # Get the 0-based indices (bit positions) where bits are active
    active_indices = np.where(active_bits_mask)[0]

    # Convert 0-based indices to 1-based port numbers and return as a list
    active_ports = (active_indices + 1).tolist()

    # np.where returns sorted indices, so list is already sorted
    return active_ports


def add_interpreted_port_columns(
    events_df: pd.DataFrame,
    input_mask_col: str = "value1",
    output_mask_col: str = "value2",
    max_ports: int = 32,
) -> pd.DataFrame:
    """
    Adds 'active_inputs' and 'active_outputs' columns to a DataFrame
    by interpreting bitmask columns representing port states using NumPy.

    Operates on and returns a modified copy of the input DataFrame.

    Parameters
    ----------
    events_df : pd.DataFrame
        The DataFrame containing the parsed StateScript event data.
    input_mask_col : str, optional
        The name of the column containing the input port bitmask values,
        by default 'value1'.
    output_mask_col : str, optional
        The name of the column containing the output port bitmask values,
        by default 'value2'.
    max_ports : int, optional
        The maximum port number to check for the bitmasks, by default 32.

    Returns
    -------
    pd.DataFrame
        A copy of the input DataFrame with 'active_inputs' and 'active_outputs'
        columns added (or updated if they existed). Prints warnings if specified
        mask columns are not found.

    Raises
    ------
    TypeError
        If the input `events_df` is not a pandas DataFrame.
    """
    if not isinstance(events_df, pd.DataFrame):
        raise TypeError("Input 'events_df' must be a pandas DataFrame.")

    # Work on a copy to avoid modifying the original DataFrame
    processed_df = events_df.copy()

    # Interpret Input Ports
    if input_mask_col in processed_df.columns:
        # Convert column to numeric, coercing errors, then apply interpretation
        input_series = pd.to_numeric(
            processed_df[input_mask_col], errors="coerce"
        )
        processed_df["active_inputs"] = input_series.apply(
            lambda mask: _interpret_port_mask(mask, max_ports)
        )
    else:
        print(
            f"Warning: Input mask column '{input_mask_col}' not found in DataFrame. Skipping 'active_inputs'."
        )
        # Add empty column if it doesn't exist for consistency
        processed_df["active_inputs"] = [[] for _ in range(len(processed_df))]

    # Interpret Output Ports
    if output_mask_col in processed_df.columns:
        output_series = pd.to_numeric(
            processed_df[output_mask_col], errors="coerce"
        )
        processed_df["active_outputs"] = output_series.apply(
            lambda mask: _interpret_port_mask(mask, max_ports)
        )
    else:
        print(
            f"Warning: Output mask column '{output_mask_col}' not found in DataFrame. Skipping 'active_outputs'."
        )
        # Add empty column if it doesn't exist for consistency
        processed_df["active_outputs"] = [[] for _ in range(len(processed_df))]

    return processed_df
```
